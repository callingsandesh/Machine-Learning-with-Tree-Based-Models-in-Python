```python
import numpy as np
import pandas as pd
```

## Importing Dataset

```python
data = pd.read_csv('indian_liver_patient/indian_liver_patient_preprocessed.csv')
```

```python
data
```

```python
data.drop('Unnamed: 0',inplace=True,axis=1)
```

## SELECTING INDEPENDENT VARIABLES AND DEPENDENT VARIABLE

```python
X=data.drop('Liver_disease',axis=1)
y=data['Liver_disease']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

## Model

while using bootstraping in ensemble , on average 63% of the instances are sampled.The remaining 37% containg the OOB instance.
As, the OOB instance are not seen by the models these can be use to test the performance of the model.
So, which evaluating the model we can use the OOB sample to test the accuracy without the need of cross_validation.

These technique is known as the OOB evaluation.

```python
# Import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

# Import BaggingClassifier
from sklearn.ensemble import BaggingClassifier

# Instantiate dt
dt = DecisionTreeClassifier(min_samples_leaf=8, random_state=1)

# Instantiate bc
bc = BaggingClassifier(base_estimator=dt, 
            n_estimators=50,
            oob_score=True,
            random_state=1)X_test.shape
```

## OOB Score vs Test Set Score

```python
from sklearn.metrics import accuracy_score
```

```python
# Fit bc to the training set 
bc.fit(X_train, y_train)

# Predict test set labels
y_pred = bc.predict(X_test)

# Evaluate test set accuracy
acc_test = accuracy_score(y_test, y_pred)

# Evaluate OOB accuracy
acc_oob = bc.oob_score_

# Print acc_test and acc_oob
print('Test set accuracy: {:.3f}, OOB accuracy: {:.3f}'.format(acc_test, acc_oob))
```

The test set accuracy and the OOB accuracy of bc are both roughly equal to 70%!
